```python
# ============================================================
#  DecodeLabs Internship - Artificial Intelligence
#  Project 3: AI Recommendation Logic
#  Method: Content-Based Filtering (TF-IDF + Cosine Similarity)
#  Capstone: Tech Stack Recommender
#  Batch: 2026
# ============================================================

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# ── KNOWLEDGE BASE: Job Roles Dataset (raw_skills) ──────────
# Each role has a tag string representing its required skills
job_roles = [
    {
        "title": "Data Scientist",
        "skills": "python sql machine learning data analysis statistics numpy pandas scikit-learn visualization"
    },
    {
        "title": "Machine Learning Engineer",
        "skills": "python machine learning deep learning tensorflow pytorch scikit-learn model deployment algorithms"
    },
    {
        "title": "AI Research Scientist",
        "skills": "python deep learning neural networks pytorch research mathematics statistics nlp computer vision"
    },
    {
        "title": "DevOps Engineer",
        "skills": "aws docker kubernetes linux ci cd git automation cloud infrastructure bash"
    },
    {
        "title": "Cloud Architect",
        "skills": "aws cloud azure google cloud kubernetes docker networking security automation infrastructure"
    },
    {
        "title": "Backend Developer",
        "skills": "python java sql apis rest django flask databases postgresql mongodb git"
    },
    {
        "title": "Frontend Developer",
        "skills": "html css javascript react nodejs typescript ui ux design web responsive"
    },
    {
        "title": "Full Stack Developer",
        "skills": "html css javascript react nodejs python django sql rest apis git"
    },
    {
        "title": "Data Engineer",
        "skills": "python sql spark hadoop etl pipeline aws data warehouse postgresql kafka big data"
    },
    {
        "title": "Cybersecurity Analyst",
        "skills": "networking linux security firewall penetration testing encryption python bash monitoring"
    },
    {
        "title": "NLP Engineer",
        "skills": "python nlp deep learning transformers bert huggingface text classification pytorch tensorflow"
    },
    {
        "title": "Computer Vision Engineer",
        "skills": "python computer vision deep learning opencv pytorch tensorflow image classification cnn"
    },
    {
        "title": "Business Intelligence Analyst",
        "skills": "sql data analysis visualization power bi tableau excel statistics reporting dashboards"
    },
    {
        "title": "Systems Administrator",
        "skills": "linux bash automation networking cloud aws docker monitoring scripting infrastructure security"
    },
    {
        "title": "Robotics Engineer",
        "skills": "python c++ ros robotics automation algorithms computer vision deep learning sensors"
    },
]

# ── PIPELINE: Build TF-IDF Matrix from Job Role Skills ──────
all_skill_docs = [role["skills"] for role in job_roles]
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(all_skill_docs)

# ── CORE FUNCTION: The 4-Step Ranking Pipeline ──────────────
def get_recommendations(user_skills: list, top_n: int = 3):
    """
    Step 1 - Ingestion  : Accept user skill inputs
    Step 2 - Scoring    : Compute Cosine Similarity scores
    Step 3 - Sorting    : Rank by descending score
    Step 4 - Filtering  : Return Top-N results only
    """
    # Step 1: Build user profile string from inputs
    user_profile = " ".join(user_skills).lower()

    # Step 2: Transform user profile into same TF-IDF vector space
    user_vector = vectorizer.transform([user_profile])

    # Step 3: Compute cosine similarity between user & all job roles
    scores = cosine_similarity(user_vector, tfidf_matrix).flatten()

    # Step 4: Sort descending, filter Top-N
    ranked_indices = np.argsort(scores)[::-1][:top_n]

    results = []
    for idx in ranked_indices:
        results.append({
            "rank": len(results) + 1,
            "title": job_roles[idx]["title"],
            "score": scores[idx],
            "skills": job_roles[idx]["skills"]
        })
    return results

# ── MAIN: Interactive Tech Stack Recommender ────────────────
def run_recommender():
    print("=" * 55)
    print("   DecodeLabs AI - Project 3")
    print("   AI Recommendation Logic")
    print("   Tech Stack Recommender (Content-Based Filtering)")
    print("=" * 55)
    print("\n💡 This engine maps your skills to the best-fit")
    print("   career paths using TF-IDF + Cosine Similarity.\n")

    print("─" * 55)
    print("📌 Enter at least 3 skills (comma-separated).")
    print("   Example: python, machine learning, sql")
    print("─" * 55)

    # ── STEP 1: Ingestion — capture user state ───────────────
    while True:
        raw = input("\nYour Skills: ").strip()
        if not raw:
            print("⚠️  Please enter at least 3 skills.")
            continue

        user_skills = [s.strip() for s in raw.split(",") if s.strip()]

        if len(user_skills) < 3:
            print(f"⚠️  You entered {len(user_skills)} skill(s). Please enter at least 3.")
            continue
        break

    print(f"\n✅ Skills captured: {user_skills}")
    print(f"\n⚙️  Running similarity engine...")
    print(f"   Method  : TF-IDF Vectorization + Cosine Similarity")
    print(f"   Dataset : {len(job_roles)} Job Roles")
    print(f"   Output  : Top 3 Career Path Recommendations\n")

    # ── STEPS 2-4: Score → Sort → Filter ────────────────────
    recommendations = get_recommendations(user_skills, top_n=3)

    # ── OUTPUT: Display Ranked Results ───────────────────────
    print("=" * 55)
    print("🏆 YOUR TOP 3 RECOMMENDED CAREER PATHS")
    print("=" * 55)

    medals = ["🥇", "🥈", "🥉"]
    for rec in recommendations:
        match_pct = rec["score"] * 100
        bar_len = int(match_pct / 5)
        bar = "█" * bar_len + "░" * (20 - bar_len)
        print(f"\n{medals[rec['rank']-1]}  Rank {rec['rank']}: {rec['title']}")
        print(f"   Match Score : [{bar}] {match_pct:.1f}%")
        print(f"   Key Skills  : {rec['skills'][:60]}...")

    print("\n" + "=" * 55)
    print("💡 Tip: Add more specific skills for better matches!")
    print("   (e.g. 'pytorch', 'kubernetes', 'postgresql')")
    print("=" * 55)

    # ── Try Again Option ─────────────────────────────────────
    again = input("\n🔄 Try with different skills? (yes/no): ").strip().lower()
    if again in ["yes", "y"]:
        run_recommender()
    else:
        print("\n✅ Project 3 Complete! Recommendation engine built.")
        print("   Thank you for using DecodeLabs Tech Stack Recommender! 🚀")

# ── ENTRY POINT ──────────────────────────────────────────────
if __name__ == "__main__":
    run_recommender()
```
