```python
# ============================================================
#  DecodeLabs Internship - Artificial Intelligence
#  Project 4: AI Text Generation Using NLP
#  Method: N-Gram Language Model (Bigram + Trigram)
#  Core: Statistical Pattern Learning & Probabilistic Reasoning
#  Batch: 2026
# ============================================================

import re
import random
import collections

# ── TRAINING CORPUS ──────────────────────────────────────────
# The model learns language patterns from this text
CORPUS = """
Artificial intelligence is transforming the world of technology.
Machine learning enables computers to learn from data and improve over time.
Deep learning uses neural networks to solve complex problems in AI.
Natural language processing helps machines understand and generate human language.
Python is the most popular programming language for artificial intelligence.
Data science combines statistics machine learning and programming to extract insights.
Neural networks are inspired by the structure of the human brain.
Algorithms are step by step instructions that computers follow to solve problems.
Artificial intelligence is used in healthcare finance education and robotics.
Machine learning models are trained on large datasets to make accurate predictions.
Deep learning has revolutionized computer vision and speech recognition.
Natural language processing powers chatbots virtual assistants and translation systems.
Python libraries like numpy pandas and scikit-learn simplify data analysis.
Data engineers build pipelines that collect clean and store large amounts of data.
Reinforcement learning trains agents to make decisions by rewarding correct actions.
Computer vision enables machines to interpret and understand visual information.
Cloud computing provides scalable infrastructure for deploying AI applications.
Automation reduces repetitive tasks and increases efficiency in many industries.
The future of artificial intelligence includes autonomous vehicles and smart cities.
Ethics in artificial intelligence ensures that technology is fair transparent and safe.
Programming is the foundation of all software and artificial intelligence systems.
Statistics and mathematics are essential skills for any data scientist or AI engineer.
Transformers and attention mechanisms have advanced natural language processing significantly.
Open source tools and frameworks accelerate the development of artificial intelligence.
Developers use version control systems like git to manage and collaborate on code.
Supervised learning uses labeled data to train models for classification and regression.
Unsupervised learning finds hidden patterns in data without the use of labels.
Transfer learning allows models to apply knowledge gained from one task to another.
Generative models can create new images text and audio by learning data distributions.
The internet of things connects physical devices to the digital world using sensors.
"""

# ============================================================
#  PHASE 1: TOKENIZATION & PREPROCESSING
# ============================================================
def preprocess(text: str) -> list:
    """Clean and tokenize the corpus into a word list."""
    text = text.lower()
    text = re.sub(r"[^a-z\s]", "", text)   # Remove non-alpha characters
    tokens = text.split()
    return tokens

# ============================================================
#  PHASE 2: N-GRAM MODEL BUILDING
# ============================================================
def build_ngram_model(tokens: list, n: int) -> dict:
    """
    Build an N-Gram frequency model.
    Key   = (n-1) preceding words (context)
    Value = list of words that can follow that context
    """
    model = collections.defaultdict(list)
    for i in range(len(tokens) - n + 1):
        context = tuple(tokens[i: i + n - 1])   # preceding words
        next_word = tokens[i + n - 1]            # word that follows
        model[context].append(next_word)
    return dict(model)

# ============================================================
#  PHASE 3: TEXT GENERATION (Probabilistic Reasoning)
# ============================================================
def generate_text(model: dict, seed_words: list, length: int = 20, n: int = 3) -> str:
    """
    Generate text using the N-Gram model.
    - Start from seed words
    - At each step, look up the context and pick the next word probabilistically
    - Continue until desired length is reached
    """
    output = list(seed_words)
    context_size = n - 1

    for _ in range(length):
        context = tuple(output[-context_size:])

        # Look up context in model
        if context in model:
            next_word = random.choice(model[context])   # Probabilistic selection
        else:
            # Fallback: try shorter context (bigram)
            short_context = (output[-1],)
            bigram_model = build_ngram_model(tokens, 2)
            if short_context in bigram_model:
                next_word = random.choice(bigram_model[short_context])
            else:
                break   # No continuation found

        output.append(next_word)

    return " ".join(output)

# ============================================================
#  PHASE 4: ANALYSIS — Show Model Statistics
# ============================================================
def show_stats(tokens: list, bigram_model: dict, trigram_model: dict):
    print(f"\n📊 Language Model Statistics")
    print(f"   Total Tokens (Words)   : {len(tokens)}")
    print(f"   Unique Vocabulary Size : {len(set(tokens))}")
    print(f"   Bigram  Contexts Built : {len(bigram_model)}")
    print(f"   Trigram Contexts Built : {len(trigram_model)}")
    print(f"\n   Top 10 Most Frequent Words:")
    freq = collections.Counter(tokens)
    for word, count in freq.most_common(10):
        bar = "█" * count
        print(f"   {word:20} {bar} ({count})")

# ============================================================
#  MAIN: Interactive Text Generation Engine
# ============================================================
def run_text_generator():
    print("=" * 55)
    print("   DecodeLabs AI - Project 4")
    print("   AI Text Generation Using NLP")
    print("   N-Gram Language Model (Bigram + Trigram)")
    print("=" * 55)

    # ── Step 1: Preprocess corpus ────────────────────────────
    print("\n⚙️  Phase 1: Tokenizing and preprocessing corpus...")
    global tokens
    tokens = preprocess(CORPUS)
    print(f"   ✅ Corpus tokenized → {len(tokens)} tokens")

    # ── Step 2: Build language models ───────────────────────
    print("\n⚙️  Phase 2: Building N-Gram Language Models...")
    bigram_model  = build_ngram_model(tokens, n=2)
    trigram_model = build_ngram_model(tokens, n=3)
    print(f"   ✅ Bigram  model built  → {len(bigram_model)}  unique contexts")
    print(f"   ✅ Trigram model built  → {len(trigram_model)} unique contexts")

    # ── Step 3: Show statistics ──────────────────────────────
    show_stats(tokens, bigram_model, trigram_model)

    # ── Step 4: Interactive Generation Loop ─────────────────
    print("\n" + "=" * 55)
    print("🤖 AI Text Generation Engine Ready!")
    print("=" * 55)
    print("\n📌 Instructions:")
    print("   Enter 1-2 seed words from the AI/tech domain.")
    print("   The model will generate a sentence continuation.")
    print("   Type 'quit' to exit.\n")

    available_seeds = list(set(tokens))
    print(f"   Sample seed words: {random.sample(available_seeds, 8)}\n")

    while True:
        print("─" * 55)
        seed_input = input("Enter seed word(s): ").strip().lower()

        if seed_input in ["quit", "exit", "bye"]:
            print("\n✅ Project 4 Complete! Language model built and tested.")
            print("   Thank you for using DecodeLabs Text Generator! 🚀")
            break

        if not seed_input:
            print("⚠️  Please enter at least one word.")
            continue

        seed_words = seed_input.split()

        # Validate seed words exist in vocabulary
        valid_seeds = [w for w in seed_words if w in tokens]
        if not valid_seeds:
            print(f"⚠️  Word(s) not in vocabulary. Try: {random.sample(available_seeds, 5)}")
            continue

        # Generate multiple variations
        print(f"\n🧠 Generating text from seed: '{' '.join(valid_seeds)}'\n")

        for i in range(3):
            # Use trigram model for richer generation
            generated = generate_text(trigram_model, valid_seeds, length=15, n=3)
            # Capitalize first letter, add period
            sentence = generated.capitalize() + "."
            print(f"   [{i+1}] {sentence}")

        print()

# ── ENTRY POINT ──────────────────────────────────────────────
if __name__ == "__main__":
    tokens = []   # global token list used in fallback
    run_text_generator()
```
