```python
# ============================================================
#  DecodeLabs Internship - Artificial Intelligence
#  Project 2: Data Classification Using AI
#  Algorithm: K-Nearest Neighbors (KNN)
#  Dataset: Iris Benchmark (150 samples, 3 classes, 4 features)
#  Batch: 2026
# ============================================================

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import (classification_report,
                             confusion_matrix,
                             f1_score,
                             accuracy_score)

# ── PHASE 1: INPUT — Load & Understand the Dataset ──────────
print("=" * 55)
print("   DecodeLabs AI - Project 2")
print("   Data Classification Using AI (KNN)")
print("=" * 55)

iris = load_iris()
X = iris.data        # Features: sepal length/width, petal length/width
y = iris.target      # Labels:   0=Setosa, 1=Versicolor, 2=Virginica

print(f"\n📊 Dataset Overview")
print(f"   Total Samples  : {X.shape[0]}")
print(f"   Features       : {X.shape[1]}  {list(iris.feature_names)}")
print(f"   Classes        : {len(iris.target_names)}  {list(iris.target_names)}")

# ── PHASE 2: PROCESS — Gatekeeper Rule: Feature Scaling ─────
scaler = StandardScaler()          # Mean=0, Variance=1
X_scaled = scaler.fit_transform(X)

print(f"\n⚙️  Preprocessing")
print(f"   StandardScaler applied → Mean=0, Variance=1")

# ── Structural Integrity: Train-Test Split (80 / 20) ────────
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y,
    test_size=0.2,
    random_state=42,
    shuffle=True          # Remove order bias
)

print(f"   Train Set Size : {len(X_train)} samples (80%)")
print(f"   Test  Set Size : {len(X_test)}  samples (20%)")

# ── The Algorithm: K-Nearest Neighbors ──────────────────────
print(f"\n🤖 Training KNN Model  (K=5, Majority Vote)")

model = KNeighborsClassifier(n_neighbors=5)   # Instantiate
model.fit(X_train, y_train)                   # Fit (Memorize the map)
predictions = model.predict(X_test)           # Predict (Apply logic)

print(f"   Model trained successfully!")

# ── PHASE 3: OUTPUT — Validation ────────────────────────────
accuracy = accuracy_score(y_test, predictions)
f1 = f1_score(y_test, predictions, average='weighted')

print(f"\n📈 Output Validation")
print(f"   Accuracy Score : {accuracy * 100:.2f}%")
print(f"   F1 Score       : {f1:.4f}  (Weighted Harmonic Mean of Precision & Recall)")

# ── Diagnostic Tool: Confusion Matrix ───────────────────────
cm = confusion_matrix(y_test, predictions)
print(f"\n🔍 Confusion Matrix")
print(f"   Rows = Actual Class | Cols = Predicted Class")
print(f"   {'':15} Setosa   Versicolor  Virginica")
for i, row in enumerate(cm):
    print(f"   {iris.target_names[i]:15} {row}")

# ── Full Classification Report ───────────────────────────────
print(f"\n📋 Full Classification Report")
print(classification_report(y_test, predictions,
                             target_names=iris.target_names))

# ── Interactive Prediction ───────────────────────────────────
print("=" * 55)
print("🌸 Try a Live Prediction!")
print("   Enter 4 flower measurements to classify it.")
print("   (Press Enter to use a sample: 5.1, 3.5, 1.4, 0.2)\n")

try:
    raw = input("   Enter [sepal_length, sepal_width, petal_length, petal_width]: ").strip()

    if raw == "":
        features = [5.1, 3.5, 1.4, 0.2]   # Classic Setosa sample
        print(f"   Using sample values: {features}")
    else:
        features = [float(x.strip()) for x in raw.split(",")]

    features_scaled = scaler.transform([features])
    result = model.predict(features_scaled)
    flower = iris.target_names[result[0]]
    print(f"\n   🌺 Prediction: This flower is → {flower.upper()}")

except Exception:
    print("   (Skipping live prediction — invalid input)")

print("\n✅ Project 2 Complete! Model trained, tested, and validated.")
print("=" * 55)
```
